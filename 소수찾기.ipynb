**문제 설명**

1부터 입력받은 숫자 n 사이에 있는 소수의 개수를 반환하는 함수, solution을 만들어 보세요.

소수는 1과 자기 자신으로만 나누어지는 수를 의미합니다.
(1은 소수가 아닙니다.)

**제한 조건**

n은 2이상 1000000이하의 자연수입니다.

**입출력 예**

#1 n = 10
- 1부터 10 사이의 소수는 [2,3,5,7] 4개가 존재하므로 4를 반환

#2 n = 5
- 1부터 5 사이의 소수는 [2,3,5] 3개가 존재하므로 3를 반환

# 나머지 계산으로 소수를 판별하여 개수 세기

소수 : 자기 자신과 1로만 나누어떨어지는 수

1은 소수가 아니므로 2부터 n까지의 수 중에서 해당 숫자보다 작은 수(2~n-1)로 나누었을 때 나머지가 0인 경우가 한번도 없으면 소수로 판단

```python
n = 100
answer = 0
prime = []
total_cnt = 0
for i in range(2, n + 1):
    # i가 소수인지 판단
    flag = 0
    print('number =', i)
    for j in range(2, i):
        print(j)
        total_cnt += 1
        if i % j == 0:
            flag = 1
            break
    if flag == 0:
        prime.append(i)
        answer += 1
        print(prime)
print('answer : %d, cnt : %d' % (answer, total_cnt))
```

```python
def solution(n):
    answer = 0
    for i in range(2, n + 1):
        flag = 0
        for j in range(2, i):
            # 약수가 발견되면 더 이상 나눗셈하지 않는다
            if i % j == 0:
                flag = 1
                break
        if flag == 0:
            answer += 1
    return answer
```

# 나머지계산으로 소수를 판별하여 리스트에 추가 (리스트에 있는 소수로만 나눗셈)

2로 나눴을 때 나머지가 0이면 2의 배수로 나누어도 나머지가 0이다

즉, n보다 작은 소수로만 나누어서 판단해도 된다 -> 이미 prime 리스트에 추가된 소수를 활용

```python
n = 100
answer = 0
prime = []
total_cnt = 0
for i in range(2, n + 1):
    # i가 소수인지 판단
    flag = 0
    print('number =', i)
    for j in prime:
        print(j)
        total_cnt += 1
        if i % j == 0:
            flag = 1
            break
    if flag == 0:
        prime.append(i)
        answer += 1
        print(prime)
print('answer : %d, cnt : %d' % (answer, total_cnt))
```

```python
def solution(n):
    prime = []
    for i in range(2, n + 1):
        flag = 0    # 소수 판별
        for j in prime:
            # 약수가 발견되면 더 이상 나눗셈하지 않는다
            if i % j == 0:
                flag = 1
                break
        if flag == 0:
            prime.append(i)
    return len(prime)
```

# 2의배수, 3의배수...(n의 반까지만 진행) 소수가 아닌 것 찾아서 지우기

소수를 찾아서 추가하는 것이 아닌 소수 판별 대상인 2부터 n까지의 수 중에서 소수가 아닌 것을 지운다
- num :  2 ~ n까지의 숫자 리스트
- 소수의 배수들을 모두 찾아 지운다
- 지우고 남은 수 중 맨 앞에 있는 수가 다음 소수
- n/2 이상의 수는 배수가 n을 넘어가기 때문에 지울 수가 없으므로, n/2 차례가 된 경우 남은 수 모두 소수

```python
n = 100
num = list(range(2, n+1))
print(num)
prime = []
total_cnt = 0

while len(num) > 0:
    tmp = num[0]
    print('prime :', tmp)
    if tmp > n / 2:
        prime.extend(num)
        break
    prime.append(num[0])
    
    i = 1
    while tmp * i <= n:
        if tmp * i in num:
            num.remove(tmp * i)
            print('delete', tmp * i)
        total_cnt += 1
        i += 1
    print('num :', num, len(num), '개')
    print()
print('-------------------------------------------------------------------------')
print(prime)
print('answer : %d, cnt : %d' % (len(prime), total_cnt))
```

```python
def solution(n):
    num = list(range(2, n + 1))
    prime = []
    
    while len(num) > 0:
        tmp = num[0]
        if tmp > n / 2:
            prime.extend(num)
            break
        prime.append(num[0])
        i = 1
        while tmp * i <= n:
            if tmp * i in num:
                num.remove(tmp * i)
            i += 1
    
    return len(prime)
```

```python
n = 100
solution(n)
```

# 2의배수, 3의배수...(n의 반까지만 진행) 소수가 아닌 것 찾아서 지우기 (ver2)

소수의 배수 중에서도 다른 소수의 배수와 겹치는 것들 존재
- 5의 경우 5*2, 5*3, 5*4는 이미 전의 소수 배수에서 지워짐
- 따라서 5*5부터 지우면 됨

```python
n = 100
num = list(range(2, n+1))
print(num)
prime = []
total_cnt = 0

while len(num) > 0:
    tmp = num.pop(0)
    prime.append(tmp)
    print('prime :', tmp)
    if tmp * tmp > n:
        prime.extend(num)
        break
    
    i = 0
    while tmp * (tmp + i) <= n:
        if tmp * (tmp + i) in num:
            num.remove(tmp * (tmp + i))
            print('delete', tmp * (tmp + i))
        total_cnt += 1
        i += 1
    print('num :', num, len(num), '개')
    print()
print('-------------------------------------------------------------------------')
print(prime)
print('answer : %d, cnt : %d' % (len(prime), total_cnt))
```

```python
def solution(n):
    num = list(range(2, n + 1))
    prime = []
    
    while len(num) > 0:
        tmp = num.pop(0)
        prime.append(tmp)
        if tmp * tmp > n:
            prime.extend(num)
            break

        i = 0
        while tmp * (tmp + i) <= n:
            if tmp * (tmp + i) in num:
                num.remove(tmp * (tmp + i))
            i += 1
    
    return len(prime)
```

```python
n = 100
solution(n)
```

# set - difference 활용

```python
n = 100
num = list(range(2, n+1))
print(num)
prime = []

while len(num) > 0:
    tmp = num[0]
    print('prime :', tmp)
    if tmp * tmp > n:
        prime.extend(num)
        break
    prime.append(tmp)
    
    prime_multiple = list(range(tmp, n+1, tmp))
    print('delete', prime_multiple)
    num = sorted(list(set(num).difference(prime_multiple)))
    print('n :', len(num), 'num :', num)
    print()
print('-------------------------------------------------------------------------')
print(prime)
print('answer : %d' % (len(prime)))
```

```python
def solution(n):
    num = list(range(2, n + 1))
    prime = []
    
    while len(num) > 0:
        tmp = sorted(list(num))[0]
        if tmp * tmp > n:
            prime.extend(num)
            break
        prime.append(tmp)
        
        prime_multiple = list(range(tmp, n+1, tmp))
        num = set(num).difference(prime_multiple)
    
    return len(prime)
```

```python
n = 100
solution(n)
```
